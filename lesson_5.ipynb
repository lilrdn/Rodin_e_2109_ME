Метод градиентного спуска для функций одной и двух перемнных

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
def f_1(x):
    return 3*(x**2)

def df_1(x):
    return 6*x
```

```python
x0 = 5
la = 0.01
res = []
x1 = x0 - la*df_1(x0)
while abs(x0-x1) > 0.00000001:
    res.append((x0, f_1(x0)))
    x0 = x1
    x1 = x0 - la*df_1(x0)
```

```python
res
```

```python
x_1 = [i[0] for i in res]
y_1 = [i[1] for i in res]
x = [i for i in range(-5, 5)]
y = [f_1(i) for i in range(-5, 5)]
plt.plot(x, y)
plt.plot(x_1, y_1)
```

```python
def f_2(xy):
    return 3*(xy[0]**2) + 5*(xy[1]**2)

def df_2(xx):
    return np.array((6*xx[0], 10*xx[1]))
```

```python
x0 = (5, 10)
la = 0.01
res = []
x1 = x0 - la*df_2(x0)
while np.linalg.norm(df_2(x1)) > 0.000001:
    res.append((x0))
    x0 = x1
    x1 = x0 - la*df_2(x0)
res
```

```python
x_1 = np.array([point[0] for point in res[1:]])
y_1 = np.array([point[1] for point in res[1:]])
```

```python
x = np.array([i for i in range(-5, 5)])
y = np.array([i for i in range(-5, 5)])
x, y = np.meshgrid(x, y)
z = np.array([f_2((x,y)) for x,y in list(zip(x,y))])

z
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Строим поверхность
ax.plot_surface(x, y, z, cmap='viridis')
ax.plot(x_1, y_1, zs=0, color='r')

# Настройка отображения
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

def update_graph(num):
    ax.view_init(elev=200., azim=num)
ani = FuncAnimation(fig, update_graph, frames=360, interval=50)
plt.show()
```

Метод наискорейшего градиентного спуска?

```python
def gradient_descent(f, df, x0, learning_rate=0.01, epsilon=1e-6, max_iter=1000):
    x = x0
    x_history = [x.copy()]
    for i in range(max_iter):
        gradient = df(x)
        if np.linalg.norm(gradient) < epsilon:
            break
        x = x - learning_rate * gradient
        x_history.append(x.copy())
    return x, np.array(x_history)

def f(x):
    return x[0]**2 + x[1]**2

def df(x):
    return np.array([2*x[0], 2*x[1]])

x0 = np.array([1.0, 1.0])
result, x_history = gradient_descent(f, df, x0)

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f([X, Y])

plt.figure()
cp = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(cp)
plt.plot(x_history[:, 0], x_history[:, 1], '-o', color='red')
plt.title('Градиентный спуск')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

print("Минимум функции:", result)
```
