```python
X = df.drop(['target', 'intercept'], axis=1).values.tolist()
y = df['target'].values.tolist()

def OLS(X, y):
    n = len(X)
    num_X = len(X[0])
    sum_x = [sum([row[i] for row in X]) for i in range(num_X)]
    sum_xy = [sum([X[j][i] * y[j] for j in range(n)]) for i in range(num_X)]
    beta = [round(xy / x, 2) if x != 0 else 0 for xy, x in zip(sum_xy, sum_x)]
    return beta

coefficients_ols = OLS(X, y)
print(coefficients_ols)
```
