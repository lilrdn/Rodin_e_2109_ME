**Задача 4**\
Сгенерировать массив из 20 случайных элементов, элементы которого лежат в интервале [0, 10].
+ Определить количество элементов, отличающихся от минимального на 5.
+ Найти среднее арифметическое всех четных элементов массива, стоящих на нечетных местах.

```python
arr = np.random.randint(0,11,20)
print(len(arr[abs(min(arr) - arr) == 5]))
print(np.mean(arr[arr%2==0][::2]))
```

**Задача 5**\
Сгенерировать матрицу A размером 5 на 5, элементы которой лежат в интервале [1, 10].
+ Возвести в квадрат максимальный элемент матрицы.
+ Все четные числа заменить на 0.

```python
arr = np.random.randint(1,11,25).reshape(5,5)
maxarr = np.max(arr)
arrmax = np.where(arr == maxarr, arr**2, arr)
print(arrmax)
arrchet = np.where(arr%2==0, 0, arr)
print(arrchet)
```

**Задача 6**\
Сгенерировать матрицу B размером 5 на 5, элементы которой лежат в интервале [-5, 5]. Отрицательные элементы матрицы заменить на число отрицательных элементов в строке, в которой они находятся.

```python
arr = np.random.randint(-5,5,25).reshape(5,5)
print(arr)
length = [len(arr[i][arr[i] < 0]) for i in range(arr.shape[0])]
for i in range(arr.shape[0]):
    arr[i][arr[i] < 0] = length[i]
print(arr)
```

**Задание 1**\
Проверить являются ли выборки связными.

```python
#проверим корреляцию между признаками по двум группам
df.loc[df['School_Type'] == 'Lib Arts'].corr()
```

```python
df.loc[df['School_Type'] == 'Univ'].corr()
```

Самый сильно коррелирующий с другими признаками - это процент принятых студентов, в теории разумно строить модель регрессионой зависимости между ним и другими признакими (например с процентом учеников из топ 10 выпускных классов)

```python
#Проверим связанность выборок по видам заведений по критерию % принятых абитуриентов по критерию Уилкоксона
#НО - процент принятых студентов в обоих видах школ одинаков
#H1 - процент принятых студентов в обоих видах заведений неодинаков
import scipy.stats as stats

stats.wilcoxon(df.loc[df['School_Type'] == 'Lib Arts']['Acceptance'], df.loc[df['School_Type'] == 'Univ']['Acceptance'])
```

Отклоняем нлевую гипотезу, следовательно с вероятностью 0,95 процент принятых абитуриентов в двух группах различается
